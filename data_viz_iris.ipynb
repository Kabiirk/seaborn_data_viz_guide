## **22. Spider Graph**

```
df = iris.groupby('Species',as_index=False).mean()
df.reset_index()
print(df.head())
```

```
#DO something for this !
labels=np.array(['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm'])
stats0 = df.loc[0,labels].values
stats1 = df.loc[1,labels].values
stats2 = df.loc[2,labels].values
```

```
#Iris-setosa
angles=np.linspace(0, 2*np.pi, len(labels), endpoint=False)
# close the plot
stats=np.concatenate((stats0,[stats0[0]]))
angles=np.concatenate((angles,[angles[0]]))

fig=plt.figure()
ax = fig.add_subplot(111, polar=True)
ax.plot(angles, stats, '', linewidth=0)
ax.fill(angles, stats, alpha=0.55)
ax.set_thetagrids(angles * 180/np.pi, labels)
ax.set_title(df["Species"][0])
#ax.grid(True)


#Iris-versicolor
angles=np.linspace(0, 2*np.pi, len(labels), endpoint=False)
# close the plot
stats=np.concatenate((stats1,[stats1[0]]))
angles=np.concatenate((angles,[angles[0]]))

fig=plt.figure()
ax = fig.add_subplot(111, polar=True)
ax.plot(angles, stats, '', linewidth=0)
ax.fill(angles, stats, alpha=0.55)
ax.set_thetagrids(angles * 180/np.pi, labels)
ax.set_title(df["Species"][1])
#ax.grid(True)


#Iris-virginica
angles=np.linspace(0, 2*np.pi, len(labels), endpoint=False)
# close the plot
stats=np.concatenate((stats2,[stats2[0]]))
angles=np.concatenate((angles,[angles[0]]))

fig=plt.figure()
ax = fig.add_subplot(111, polar=True)
ax.plot(angles, stats, '', linewidth=0)
ax.fill(angles, stats, alpha=0.55)
ax.set_thetagrids(angles * 180/np.pi, labels)
ax.set_title(df["Species"][2])
#ax.grid(True)
```

## **23. Donut Plot**

## **25. Cluster Map (dendogram and tree)**

```
df = iris.loc[:,["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm"]]
df1 = iris.Species
x = dict(zip(df1.unique(),"rgb"))
row_colors = df1.map(x)
cg = sns.clustermap(df,row_colors=row_colors,figsize=(12, 12),metric="correlation")
plt.setp(cg.ax_heatmap.yaxis.get_majorticklabels(),rotation = 0,size =8)
plt.show()
```

## **28. Stacked Histogram**

```
iris['Species'] = iris['Species'].astype('category')

list1=list()
mylabels=list()
for gen in iris.Species.cat.categories:
    list1.append(iris[iris.Species==gen].SepalLengthCm)
    mylabels.append(gen)
    
h=plt.hist(list1,bins=30,stacked=True,rwidth=1,label=mylabels)
plt.legend()
plt.show()
```
